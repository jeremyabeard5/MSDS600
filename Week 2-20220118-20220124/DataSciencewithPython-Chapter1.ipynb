```python
#now we define a list with all names of categorical features in data. also we loop through every variable in the list, getting dummy variable encoded output. 
cat_vars = data_column_category
for var in cat_vars:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(df[var], prefix=var)
    data1=df.join(cat_list)
    df=data1
df.columns
```

```python
# now we neglect the categorical column for which we ahve done encoding. we'll select only the numerical and encoded categorical columns
cat_vars = data_column_category
data_vars = df.columns.values.tolist()
to_keep=[]
for i in data_vars:
    if i not in cat_vars:
        to_keep.append(i)
        
#selecting only the numerical and encoded categorical columns
data_final=df[to_keep]
data_final.columns
```
